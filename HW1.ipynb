```python
import random
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['axes.unicode_minus'] = False 

#generate a data set of 20. 
#for simplicity, 10 in the first quadrant, another 10 in the third quadrant 
X1 = []
Y1 = []
X2 = []
Y2 = []

for i in range(10):
    X1.append(random.uniform(0,1))
    Y1.append(random.uniform(0,1))
    X2.append(random.uniform(-1,0))
    Y2.append(random.uniform(-1,0))
    
#label the data
data1 = [np.array([1,X1[i],Y1[i],1]) for i in range(10)]
data2 = [np.array([1,X2[i],Y2[i],-1]) for i in range(10)]
data = data1 + data2
```

```python
%matplotlib inline
rng = np.random.RandomState(1)
w = rng.randn(3,1)
x1 = -1
y1 = -1 / w[2] * (w[0] * 1 + w[1] * x1)
x2 = 1
y2 = -1 / w[2] * (w[0] * 1 + w[1] * x2)
plt.plot([x1,x2], [y1,y2])
plt.plot(X1, Y1, '*')
plt.plot(X2, Y2, '.')
plt.xlabel('x')
plt.ylabel('y')
```

```python
data = np.array(data)
for i in range(100):
    s = np.dot(data[:, 0:3], w)
    y_pred = np.ones((len(data), 1))
    loc_n = np.where(s < 0)[0]
    y_pred[loc_n] = -1
    num_fault = len(np.where(data[:,-1] != y_pred.T)[0])
#     print('Iteration %2d, wrongly classified points：%2d' % (i, num_fault))
    if num_fault == 0:
        break
    else:
        t = np.where(data[:,-1].reshape(len(data),1) != y_pred)[0][0]
        w += data[t][-1] * data[t, :-1].reshape((3,1))
print('%2d iterations to converge.' %(i))
```
